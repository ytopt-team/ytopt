```r
Autotuning the application using ytopt-libe (4 workers, 3 parameters)
```

```r
library(doMC)
#Register multiple cores
registerDoMC(4)

set.seed(3456)
mdata <- read.csv("results.csv")

t_orig = 4.2e-05
col_index <-  mdata[,4:6]
#print(col_index)
ll <- length(col_index[,1])
avg <- col_index[1,]
min <- min(col_index[,1])
print('maximum: ')
print(min)
print('Improvement percentage(%): ')
print(100*(min-t_orig)/t_orig)

x<- c(1,max(col_index[,2]))
y<-  t_orig * x/x

plot(col_index[,2], col_index[,1], col="blue", type="b",cex.lab=1.3,cex.axis=1.2,lwd=2, xlab ="Wall-clock Time (s)", ylab = "MSE Difference",  main ="Autotuning Mix-Kernels-Spline using ytopt-libe")
lines(col_index[,2], col_index[,3], cex.lab=1.3,cex.axis=1.2,lwd=2,col="green", type="o")
lines(x, y, cex.lab=1.3,cex.axis=1.2,lwd=2,col="red", type="l")
grid(nx = NULL, ny = NULL, col = "black", lty = "dotted")
legend(4500,0.05,c("baseline","autotuning","Best"),lwd=2, col=c("red","blue","green"),cex=1.2)
```
